```python
import random
import pandas as pd
import numpy as np
```

```python
# Create lists of random data
loan_amount = [random.randint(100000, 500000) for i in range(1000)]
credit_score = [random.randint(640, 850) for i in range(1000)]
property_value = [random.randint(200000, 700000) for i in range(1000)]
loan_to_value = [random.uniform(0.5, 1.0) for i in range(1000)]
loan_term = [random.choice([15, 20, 25, 30, 40]) for i in range(1000)]
employment_length = [random.choice(['< 1 year', '1-3 years', '3-5 years', '5-10 years', '10+ years']) for i in range(1000)]
location = [random.choice(['Urban', 'Suburban', 'Rural']) for i in range(1000)]
loan_type = [random.choice(['Fixed', 'Adjustable']) for i in range(1000)]
property_type = [random.choice(['Single Family', 'Multi Family', 'Condo', 'Townhouse']) for i in range(1000)]
state = [random.choice(['CA', 'NY', 'TX', 'FL', 'IL', 'PA', 'OH', 'GA', 'NC', 'MI']) for i in range(1000)]
debt_to_income_dti= [random.randint(10, 50) for i in range(1000)]
down_payment= [random.randint(10000, 100000) for i in range(1000)]
income= [random.randint(100000, 350000) for i in range(1000)]
occupancy_type= [random.choice(['Primary', 'Secondary', 'Investment']) for i in range(1000)]
```

```python
# Calculate interest rates
interest_rate = [round(random.uniform(2.5, 5.5) + (credit_score[i]/100) - (loan_to_value[i]*0.1) - (loan_term[i]*0.2), 2) for i in range(1000)]
```

```python
# Create a dictionary of the data
data = {'Loan Amount': loan_amount,
        'Credit Score': credit_score,
        'Property Value': property_value,
        'Loan-to-Value Ratio': loan_to_value,
        'Loan Term (years)': loan_term,
        'Employment Length': employment_length,
        'Location': location,
        'Loan Type': loan_type,
        'Property Type': property_type,
        'State': state,
        'Interest Rate': interest_rate,
        'Debt-To-Income DTI':debt_to_income_dti,
        'Down Payment $': down_payment,
        'Income': income,
        'Ocuppancy Type': occupancy_type
       }
# Convert dictionary to a dataframe
df = pd.DataFrame(data)
# Print the first 10 rows of the dataframe
df.head(100)
```
